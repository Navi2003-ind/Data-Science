```python
#5: ??


dates = pd.date_range(start='2023-01-01', end='2023-01-31')

base_temp = np.linspace(10, 25, len(dates))

# Add random noise to create variation
noise = np.random.normal(0, 2, len(dates))
temperatures = base_temp + noise


df = pd.DataFrame({'Date': dates, 'Temperature': temperatures})

df_pivot = df.pivot(index='Date', columns='Temperature', values='Temperature')

sns.heatmap(df_pivot, cmap='coolwarm', annot=True)
plt.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
#SEABORN
#1:
np.random.seed(30)
data = {
    'X' : np.random.uniform(-10, 10, 300),
    'Y' : np.random.uniform(-10, 10, 300),
    'Z' : np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
fig = px.scatter_3d(df, x='X', y='Y', z='Z')
fig.show()
```

```python
#2:
np.random.seed(15)
data = {
    'Grade' : np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score' : np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
sns.violinplot(x='Grade', y='Score', data=df)
plt.show()
```

```python
#3
np.random.seed(20)
data = {
    'Month' : np.random.choice(['jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day' : np.random.choice(range(1, 31), 100),
    'Sales' : np.random.randint(1000, 10000, 100)
}
df = pd.DataFrame(data)
sales_pivot = df.pivot_table(index='Month', columns='Day', values='Sales', fill_value=0)
sns.heatmap(sales_pivot, cmap='coolwarm')
plt.show()
```

```python
#4:
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X' : x.flatten(),
    'Y' : y.flatten(),
    'Z' : z.flatten()
}
df = pd.DataFrame(data)
fig = px.scatter_3d(df, x='X', y='Y', z='Z')
fig.show()
```

```python
#5:
np.random.seed(25)
data = {
    'Country' : ['USA', 'canada', 'UK', 'Germany', 'France'],
    'Population' : np.random.randint(100, 1000, 5),
    'GDP' : np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)
fig = px.scatter(df, x='GDP', y='Population', size='Population',
                 color='Country', hover_name='Country',
                 size_max=60, title='Population vs GDP')
fig.show()
```

```python
#1:
x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2)
show(p)
```

```python
#2:
x = np.random.random(size=200) * 100
y = np.random.random(size=200) * 100
sizes = np.random.randint(5, 20, size=200)
colors = np.random.choice(["red", "blue", "green"], size=200)

p = figure(title="Scatter Plot with Random Data")

p.scatter(x, y, size=sizes, color=colors)
show(p)
```

```python
#3:
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts")
p.vbar(x=fruits, top=counts, width=0.5)

show(p)
```

```python
#4:
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins = 30)
p = figure(title="Histogram of Random Data")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)
show(p)
```
